## 2. Dropping columns

```python
data_nba.columns
```

Looking at the dataset of stats per 100 Possesions we can drop the following:
- Ranking, Age, Unnamed: 29, Games started
- Shooting Percentages since we also have columns for attempted and made shots for each category

```python
data_nba = data_nba.drop(columns=["Rk","Age", "Unnamed: 29", "GS","FG%","3P%","2P%", "FT%"])
test_poss = test_poss.drop(columns=["Rk","Age", "Unnamed: 29", "GS","FG%","3P%","2P%", "FT%"])
```

```python
shoot_nba.columns
```

Looking at the dataset of shots we can drop the following:
- Ranking, Age, Team, #
- Position, Games and Minutes played since they are already included in the other table

```python
shoot_nba = shoot_nba.drop(columns=["Age","Tm","#","Rk","Pos","G","MP","Unnamed: 9",
                                    "Unnamed: 16","Unnamed: 23", "Unnamed: 29", "Unnamed: 32",
                                    "Unnamed: 26","Att.","#.1"])
test_shoot = test_shoot.drop(columns=["Age","Tm","#","Rk","Pos","G","MP","Unnamed: 9",
                                    "Unnamed: 16","Unnamed: 23", "Unnamed: 29", "Unnamed: 32",
                                    "Unnamed: 26","Att.","#.1"])
```

## 3. Cleaning/Conatenating dataset/tables

For the shoot dataset we need to rename the columns since the initial table from basketball-reference had subheaders.

```python
shoot_nba = shoot_nba.rename(columns={"Dist.": "avg_shoot_dist", '2P': "2P_FGA", '0-3': "2P_0-3", '3-10': "2P_3-10",
 '10-16': "2P_10-16", '16-3P': "2P_16-3P", '3P': "3P_FGA", '2P.1': "2P%", '0-3.1': "2P%_0-3",
 '3-10.1': "2P%_3-10", '10-16.1': "2P%_10-16", '16-3P.1': "2P%_16-3P", '3P.1': "3P%", '2P.2': "2P%_asst",
 '3P.2': "3P%_asst", '%FGA': "Dunks%_FGA", '%3PA': "Corner3%_3PA", '3P%': "Corner3%"})
test_shoot = test_shoot.rename(columns={"Dist.": "avg_shoot_dist", '2P': "2P_FGA", '0-3': "2P_0-3", '3-10': "2P_3-10",
 '10-16': "2P_10-16", '16-3P': "2P_16-3P", '3P': "3P_FGA", '2P.1': "2P%", '0-3.1': "2P%_0-3",
 '3-10.1': "2P%_3-10", '10-16.1': "2P%_10-16", '16-3P.1': "2P%_16-3P", '3P.1': "3P%", '2P.2': "2P%_asst",
 '3P.2': "3P%_asst", '%FGA': "Dunks%_FGA", '%3PA': "Corner3%_3PA", '3P%': "Corner3%"})
```

```python
shoot_nba.shape
```

```python
data_nba.shape
```

Concatenating both tables based on the player.

```python
complete_nba = pd.concat([data_nba.set_index('Player'),shoot_nba.set_index('Player')], axis=1, join='inner').reset_index()
test_nba = pd.concat([test_poss.set_index('Player'),test_shoot.set_index('Player')], axis=1, join='inner').reset_index()
```

```python
complete_nba.head()
```

In the database of basketball-reference some players have associated 2 Positions. Since we just want to predict 5 Positions I decided to use only the first mentioned position and assign it to the player.

```python
complete_nba.Pos.value_counts()
```

```python
complete_nba['Pos'] = complete_nba['Pos'].map(lambda x: x.split('-')[0])
test_nba['Pos'] = test_nba['Pos'].map(lambda x: x.split('-')[0])
```

In the database we have Minutes played (MP) as an absolute number In order to have a comparable value across all players I will calculate the Minutes played per Game as a new column and drop the absolute value.

```python
complete_nba["MPG"] = complete_nba["MP"]/complete_nba["G"]
test_nba["MPG"] = test_nba["MP"]/test_nba["G"]
```

```python
complete_nba = complete_nba.drop(columns=["MP"])
test_nba = test_nba.drop(columns=["MP"])
```

## 4. Filtering out data

A game in the nba takes 48 minutes and 5 players are on the court at the same time. For the purpose of this model I want to make sure that I only take into account rotational players who play enough minutes per game. For this I have decided to use only players who have played more than 9 Minutes per Game. In addition a player should have played mor than 10 games in the whole season of 82 games.

```python
complete_nba.shape
```

```python
len(complete_nba.loc[(complete_nba['MPG'] >= 10) & (complete_nba['G'] > 10)])
```

```python
clean_nba = complete_nba.loc[(complete_nba['MPG'] >= 10) & (complete_nba['G'] > 10)]
test_clean_nba = test_nba.loc[(test_nba['MPG'] >= 10) & (test_nba['G'] > 10)]
```

## 5. Final Cleaning

Cecking for nulls and replacing them with 0 since they are related to players who havent shoot from a certain distance.

```python
clean_nba.isna().sum()
```

```python
def nans(df): return df[df.isnull().any(axis=1)]

nulls_clean_nba = nans(clean_nba)
```

```python
nulls_clean_nba.head()
```

```python
clean_nba = clean_nba.fillna(0)
test_clean_nba = test_clean_nba.fillna(0)
```

Clean player names from the addition after the the backslash and organize the order of columns.

```python
clean_nba['player_new'] = clean_nba['Player'].str.split('\\').str[0]
test_clean_nba['player_new'] = test_clean_nba['Player'].str.split('\\').str[0]
```

```python
clean_nba = clean_nba.drop(columns=["Player"])
test_clean_nba = test_clean_nba.drop(columns=["Player"])
```

```python
clean_nba = clean_nba[['player_new','Pos','Tm', 'G', 'MPG', 'FG', 'FGA', '3P', '3PA', '2P', '2PA', '2P%','2P%', 'FT', 'FTA',
       'ORB', 'DRB', 'TRB', 'AST', 'STL', 'BLK', 'TOV', 'PF', 'PTS', 'ORtg',
       'DRtg', 'FG%', 'avg_shoot_dist', '2P_FGA', '2P_0-3', '2P_3-10',
       '2P_10-16', '2P_16-3P', '3P_FGA', '2P%_0-3', '2P%_3-10',
       '2P%_10-16', '2P%_16-3P', '3P%', '2P%_asst', '3P%_asst', 'Dunks%_FGA',
       'Corner3%_3PA', 'Corner3%']]
test_clean_nba = test_clean_nba[['player_new','Pos','Tm', 'G', 'MPG', 'FG', 'FGA', '3P', '3PA', '2P', '2PA', '2P%','2P%', 'FT', 'FTA',
       'ORB', 'DRB', 'TRB', 'AST', 'STL', 'BLK', 'TOV', 'PF', 'PTS', 'ORtg',
       'DRtg', 'FG%', 'avg_shoot_dist', '2P_FGA', '2P_0-3', '2P_3-10',
       '2P_10-16', '2P_16-3P', '3P_FGA', '2P%_0-3', '2P%_3-10',
       '2P%_10-16', '2P%_16-3P', '3P%', '2P%_asst', '3P%_asst', 'Dunks%_FGA',
       'Corner3%_3PA', 'Corner3%']]
```

```python
clean_nba = clean_nba.loc[:,~clean_nba.columns.duplicated()]
test_clean_nba = test_clean_nba.loc[:,~test_clean_nba.columns.duplicated()]
```

Drop column Games since we only needed it for the calculation of Minutes per Gam and to filter out the players who werent really active in that season.

```python
clean_nba = clean_nba.drop(columns=["G"])
test_clean_nba = test_clean_nba.drop(columns=["G"])
```
